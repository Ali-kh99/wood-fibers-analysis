# Installation des packages nécessaires pour exécuter et visualiser ACP

```r
# Installing necessary packages to run pca
install.packages("factoextra")
install.packages("FactoMineR")
```

# Chargement et visualisation des données

```r
# Import FactoMineR Library
library('FactoMineR')

# Chargement des données
wood <- read.csv("/home/jovyan/wood-fibres.csv")

# Affichage des données
head(wood)
```

# Calcule de l'ACP
##### À l'aide de la focntion FactoMineR:PCA()

```r
# Calcule de l'ACP
wood.pca <-PCA(wood[,-1], scale.unit=TRUE, graph=F)
```

# Affichage des nuages des points par couples de variables 

```r
# Affichage des nuages des points par couples de variables 
plot(wood[,-1], pch=16, cex=0.8)
```

## Interpretations
 * D’après le graphe on peut déduire qu’il y a une corrélation entre les trois variables KinkIndex, NumberKink et TotalKinkAngle
 * Par contre on trouve qu'il ya une corrélation faible entre ces trois variable et les variable Curl et With

### Summary de notre objet ACP

```r
# summary of our PCA (wood.pca) Object
summary(wood.pca)
```

###  Voir l'objet ACP 

```r
# Appelons str() pour voir l'objet ACP 
str(wood.pca)
```

# Valeurs Propres/Variances :

```r
# import des bibliothèques 'factoextra' pour le calcul des valeurs propres et/variances 
# et 'ggplot2' pour tracer des graphes 2D
library("factoextra")
library("ggplot2")

# le calcul des valeurs propres et le % des variances 
eig.val <- get_eigenvalue(wood.pca) 
eig.val
```

### On peut visualiser ce pourcentage des variances expliquées par la fonction fviz_eig

```r
# visualisation des pourcentage des variances
fviz_eig(wood.pca, addlabels = TRUE, ylim = c(0, 70))
```

### Interpretations
* PC1 explique 61.5% de la variance totale, ce qui signifie que plus de la moitier des informations de l’ensemble de données peuvent être encapsulées par une seule composante principale.
* PC2 explique 19.8% de la variance. 
* PC3 explique 13,2% de la variance.
* Donc la réduction de l’espace de départ en trois dimensions englobe quasiment toutes les informations (94.48%) en préservant les relations existantes déjà entre les variables.

# Graphique de corélation des variables

```r
# Visualusation de variable
# Graphique de corélation des variables
fviz_pca_var(wood.pca, col.var = "blue")
```

### Interpretations

- Les variables TotalkinkAngl, Kinkindex et NumberKink presque se coïncident, car elles forment un petit angle, ce qui signifie qu’elles sont fortement corrélée positivement.

- On peut dire que la corrélation entre les deux variables Curl et With est moyenne.

- Par contre la corrélation entre les variabla TotalkinkAngl, Kinkindex et NumberKink et les 2 variables Curl et With est trés faible ou nulle.
